```python
import yfinance as yf
import pandas as pd
```

```python
ticker = "AAPL"
stock = yf.Ticker(ticker)
```

```python
data = stock.history(period="1mo")
data.head()
```

```python
print("Stock:", ticker)
print("Latest Close Price:", data['Close'][-1])
print("Average Close Price (last month):", data['Close'].mean())
```

```python
import matplotlib.pyplot as plt

data['Close'].plot(title=f"{ticker} Closing Prices - Last Month")
plt.show()
```
